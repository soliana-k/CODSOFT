```python
import pandas as pd
import numpy as np
```

```python
# i used the dataset from kaggle. the dataset has been PCA transformed already. so no feature engineering needed
train_data=pd.read_csv('/content/creditcard.csv')
train_data
```

```python
train_data.isna().sum() # i checked for missing values here. there are none
```

```python
train_data['Class'].value_counts()
# the output below shows the high imbalance between fraud and legitimate cases
```

```python
train_data.describe()
```

```python
X=train_data.drop('Class', axis=1)
y=train_data['Class']
```

```python
from sklearn.model_selection import train_test_split
train_X,val_x,train_y,val_y=train_test_split(X,y,test_size=0.2, stratify=y,random_state=0)
```

```python
from imblearn.over_sampling import SMOTE
# i applied smote because the dataset was highly imbalanced
smote=SMOTE(sampling_strategy='auto', random_state=0)
x_train_sampled,y_train_sampled=smote.fit_resample(train_X, train_y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score


## i choose random forest algorithm because it did better ( fitted better, generalized well) than logistic regression
model=RandomForestClassifier(random_state=42, class_weight='balanced')
model.fit(x_train_sampled, y_train_sampled)

y_pred=model.predict(val_x)
print("Accuracy:", accuracy_score(val_y, y_pred))
print(classification_report(val_y, y_pred))
print("ROC-AUC:", roc_auc_score(val_y, y_pred))
```

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
# i was suspicious of the accuracy score of 0.99 so i plotted the learning curve using fold of 3 for cross validation

train_sizes, train_scores, test_scores = learning_curve(model, train_X, train_y, cv=3)

plt.plot(train_sizes, train_scores.mean(axis=1), label='Train')
plt.plot(train_sizes, test_scores.mean(axis=1), label='Test')
plt.xlabel('Training Size')
plt.ylabel('ROC-AUC Score')
plt.legend()
plt.show()
```
